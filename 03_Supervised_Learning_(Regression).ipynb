# 학습 데이터 업로드

```python
#@title Target Value { display-mode: "form" }

target_value_name = 'median_house_value' #@param {type:"string"}

print(target_value_name)
```

```python
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

```python
import pandas as pd
import io
filename = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[filename]))
```

```python
df
```

파일 잘 로드했다고 알려줌

# 모델 학습

```python
# Create some data for the regression

X = df[df.columns.difference([target_value_name])]
y = df[target_value_name]
```

```python
X
```

```python
y
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# fit the regression model
model = LinearRegression()
model.fit(X, y)
```

선형회귀 모델을 만들 건데 위에 두줄은 이걸 쓸거야, 아래 두 줄은 모델 만들거야 의미

# 신규 데이터 예측 (Inference)

```python
from google.colab import files

uploaded2 = files.upload()

for fn in uploaded2.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded2[fn])))
```

```python
import io
filename = list(uploaded2.keys())[0]
df2 = pd.read_csv(io.BytesIO(uploaded2[filename]))
```

```python
df2
```

```python
# create some new points to predict
X2 = df2[df2.columns.difference([target_value_name])]

# predict the labels
y2 = model.predict(X2)
```

마지막줄 -> x2데이터로 y2데이터값을 가져올거야

```python
df2[target_value_name] = y2
```

```python
df2
```

```python
df2.to_csv('[result] '+filename)
files.download('[result] '+filename)
```
